```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=[10,12,15,18,20,22,25,30,35,40]
df = pd.DataFrame(data,columns=["values"])
print("Dataset:")
print(df)
print("\nSummary Statistics:")
print(df.describe())
print("\nvariance:",np.var(data))
print("Standard deviation:",np.std(data))

plt.hist(data)
plt.xlabel("values")
plt.ylabel("Frequency")
plt.title("Data distribution")
plt.show()

mean_value=np.mean(data)
median_value=np.median(data)
print("Inference")
print("Mean value:",mean_value)
print("Median value:",median_value)

if mean_value>median_value:
  print("Data is slightly right_skewed")
elif mean_value<median_value:
  print("Data is slightly left_skewed")
else:
  print("Data is symmetric")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
url = "https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/12_sample_data.csv"
df = pd.read_csv(url)
print("Dataset (first 10 rows):")
print(df.head(10))
data = df['Porosity']
print("\nSummary statistics:")
print(data.describe())
print("\nVariance:", np.var(data))
print("Standard deviation:", np.std(data))
mean_value = np.mean(data)
median_value = np.median(data)
plt.hist(data, bins=10, color="skyblue", edgecolor="black")
plt.axvline(mean_value, color="red", linestyle="dashed", linewidth=2,
            label=f'Mean = {mean_value:.2f}')
plt.axvline(median_value, color="green", linestyle="dashed", linewidth=2,
            label=f'Median = {median_value:.2f}')
plt.xlabel("Porosity (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Porosity")
plt.legend()
plt.show()
print("\nInference")
print("Mean value:", mean_value)
print("Median value:", median_value)
if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```

```python
# Data Manipulation
import pandas as pd

# Create DataFrame
data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 22],
    "Marks": [70, 80, 90]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
# Add new column
df["Bonus"] = 5
print("\nAfter adding Bonus:")
print(df)
# Increase marks
df["Marks"] = df["Marks"] + 2
print("\nAfter increasing marks")
print(df)
# Change value
df.loc[0, "Age"] = 21
print("\nAfter changing age of first student:")
print(df)
#rename column
df.rename(columns={'marks':'final'},inplace=True)
print('\n after renaming marks to final:')
print(df)
#remove column
df.drop('Bonus', axis=1, inplace = True)
print('\n after removing Bonus:')
print(df)
```

```python
# Selection Sort
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
print("\nOriginal array:", arr)
# Selection sort logic
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j
    # Swap the found minimum element with the first element
    arr[i], arr[min_index] = arr[min_index], arr[i]
print("\nSorted array:", arr)
```

```python
# Bubble sort
n = int(input("Enter number of elements: "))

arr = []

for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("\nOriginal array:", arr)

# Bubble sort logic
for i in range(n - 1):
    for j in range(n - 1 - i):
        if arr[j] > arr[j + 1]:
            # Swap
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print("Sorted array:", arr)
```

```python
#Linear Search

n = int(input("enter no. of elements:"))
arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i}:")))

key = int(input("enter element to search:"))

found = False

for i in range(n):
    if arr[i] == key:
        print(f"element {key} found at position {i+1}")
        found = True
        break

if not found:
    print("element not found:")
```

```python
#Binary Search

n = int(input("enter number of elements:"))

arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

arr.sort()
print("\n sorted list : arr")

key = int(input("enter element to search:"))

low = 0
high = n-1

found = False
while low <= high:
    mid = (low + high) // 2

    if arr[mid] == key:
        print(f"element {key} found at position {mid+1}")
        found = True
        break

    elif key < arr[mid]:
        high = mid - 1
    else:
      low=mid+1
    if not found:
      print ("element not found")
```

```python
# Grouping
import pandas as pd

# Simple Dataframe
data = {
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [80, 85, 75, 90, 88]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

print("\n Average Marks by department:")
print(df.groupby("Department")["Marks"].mean())

print("\nTotal marks by department:")
print(df.groupby("Department")["Marks"].sum())
```

```python
# Ranking

import pandas as pd

data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [85, 92, 78, 80]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

df["Rank"] = df["Marks"].rank(ascending=False)

print("\nAfter ranking:")
print(df)
```

```python
# Rearrange
import pandas as pd
data = {
    "Name": ["A", "B", "C"],
    "Age": [22, 21, 23],
    "Marks": [85, 90, 80]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
# Rearranging rows by Marks (ascending)
df_sorted = df.sort_values(by="Marks")
print("\nAfter rearranging rows by marks (ascending):")
print(df_sorted)
# Rearranging rows by Marks (descending)
df_sorted_desc = df.sort_values(by="Marks", ascending=False)
print("\nAfter rearranging rows by marks (descending):")
print(df_sorted_desc)
# Rearranging columns
df_rearranged_cols = df[["Marks", "Name", "Age"]]
print("\nAfter rearranging columns:")
print(df_rearranged_cols)
```

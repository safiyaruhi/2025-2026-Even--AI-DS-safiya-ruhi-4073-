```python
#21 Z-score
import numpy as np
data =[10,20,30,40,50]
mean=np.mean(data)
print("mean:",mean)
std_dev=np.std(data,ddof=1)
print("standard deviation",std_dev)
x=40
z_score=(x-mean)/std_dev
print("z-score of 40:",z_score)
z_score=[(x-mean)/std_dev for x in data]
print("z-score:",z_score)
```

```python
#21
import numpy as np
import math
population = np.array([10,20,30,40,50,60,70])
N = len (population)
print("population size(N):",N)
mu=np.mean(population)
print("population mean(u):",mu)
sigma2=np.var(population,ddof=0)
print("population variance(sigma^2):",sigma2)
sigma=np.sqrt(sigma2)
print("population standard deviation:",sigma2)
sample=np.array([20,30,40,50])
n=len(sample)
print("\n sample size(n):",n)
x_bar=np.mean(sample)
print("sample mean:",x_bar)
s2=np.var(sample,ddof=1)
print("sample variance:",s2)
s=np.sqrt(s2)
print("sample standard deviation(s)",s)
SE=sigma/math.sqrt(n)
print("standard error:",SE)
z=(x_bar-mu)/SE
print("\n z-score:",z)
```
